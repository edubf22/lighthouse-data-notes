### Task 2
Test the hypothesis that the offensive productions stats of the Cleveland Cavaliers and Golden State Warriors (the teams that met in the finals that year) were from the same distribution in the 2015/2016 season.

Offensive production refers to two variables: **PTS (Points)** and **FG_PCT (Field Goal Percentage)**. We will need to do two separate hypothesis tests, one for each variable.

```python
# data for points for each team
pts_cle = df2015[df2015['TEAM_NAME'] == 'Cleveland Cavaliers']['PTS']
fgp_cle = df2015[df2015['TEAM_NAME'] == 'Cleveland Cavaliers']['FG_PCT']
```

```python
pts_gsw = df2015[df2015['TEAM_NAME'] == 'Golden State Warriors']['PTS']
fgp_gsw = df2015[df2015['TEAM_NAME'] == 'Golden State Warriors']['FG_PCT']
```

We observe two independent samples, i.e. PTS for the GSW and CLE, and we are considering whether the two samples were drawn from the same population (e.g. distribution) or two different populations.

The t-test quantifies the difference between the arithmetic means of the two samples. The p-value quantifies the probability of observing as or more extreme values assuming the null hypothesis, that the samples are drawn from populations with the same population means, is true.

```python
# independent t-test to check if the PTS for CLE and GSW come from the same distribution
t_stat, p_val = st.ttest_ind(pts_cle, pts_gsw)
print('The t statistic is:', t_stat, '\nThe p-value is:', p_val)
```

```python
# independent t-test to check if the FG_PCT for CLE and GSW come from the same distribution
t_stat, p_val = st.ttest_ind(fgp_cle, fgp_gsw)
print('The t statistic is:', t_stat, '\nThe p-value is:', p_val)
```

```python
# what type of output do we get from st.ttest_ind?
print(type(t_stat))
```

```python
def indttest(data1,data2,nullhyp="Not specified",to_print=False,alpha=0.05):
    """Check whether or not we reject a given null hypothesis
    Parameters:
        nullhyp: (string) Null hypothesis
        data1, data2: Panda series of the 2 samples to compare.
        alpha: (float, optional) P-value threshold for significance. Default is 0.05.
        to_print: (boolean, optional) Print conclusion or not. Default is no.
    Returns p-value
    """
    # new variable that will hold the results from our independent t_test
    ttest = st.ttest_ind(data1, data2)
    
    # bit of test showing what we tested, if user enters their hypothesis
    print('Tested null hypothesis:', nullhyp)
    
    # ttest[1] gives the p-value, we will compare it to the alpha value from our indttest function
    # if p-value is less than alpha, then we reject the null hypothesis
    if ttest[1] < alpha:
        print('We rejected the null hypothesis since p-value is less than the critical value', alpha) 
        print('Samples are significantly different')
        print('p-value is:', ttest[1])
    
    # if p-value is more than alpha, then we cannot reject the null hypothesis
    else:
        print('We accept the null hypothesis since p-value is greater than the critical value', alpha)
        print('Samples are not significantly different')
        print('p-value is:', ttest[1])

indttest(pts_cle, pts_gsw, "Points are the same for the 2015/16 season.",True)
indttest(fgp_cle, fgp_gsw, "Field goal percentages are the same for the 2015/16 season.", True)
```

### Task 3
Test the hypothesis that the number of points (PTS) scored by Cleveland Cavaliers changed significantly after the head coach changed in the 2015/2016 season.

- **Coach Blatt was fired on 24th of Jan, 2016**. 

Use the data from seasons 2014/2015 and 2015/2016 - those are years when Cleveland was coached by Blatt. 

**We have two possible solutions to try here:**
- Take the same amount of games from before and after and try t-test.
- Take all the games from before and after and look for the right test to compare two samples with different sizes. (You will need to go through the scipy documentation or google to figure out what kind of test is required.)

```python
# Play around with the data to learn how to get GAME_DATE info from df2015 before Blatt was fired
df_b2015 = df2015[(df2015['TEAM_NAME'] == 'Cleveland Cavaliers') & (df2015['GAME_DATE'] < '2016-01-24')]['PTS']
df_b2015.head()
```

```python
# Now select games with coach blatt during the 2014/2015 season
df_b2014 = df2014[df2014['TEAM_NAME'] == 'Cleveland Cavaliers']['PTS']
df_b2014.head()
```

```python
# Concatenate both dfs with games under coach Blatt, pass in a list with objs, specify axis
# axis=0 means concat on rows, axis=1 means concat on columns
df_b = pd.concat([df_b2014, df_b2015], axis=0)
df_b.head()
```

```python
# now a dataframe with games without coach blatt
df_wb = df2015[(df2015['TEAM_NAME'] == 'Cleveland Cavaliers') & (df2015['GAME_DATE'] > '2016-01-23')]['PTS']
df_wb.head()
```

```python
# Since samples have different size, we need to perform a Welch's t-test
# We can use the same ttest_ind method, however we need to specify parameter equal_var=False
t_val_cle, p_val_cle = st.ttest_ind(df_b, df_wb, equal_var=False)
print('The t-statistic is:', t_val_cle, '\nThe p-value is:', p_val_cle)
```

Since the p-value is less than 0.05, we reject the null hypothesis: PTS scored per game did change after firing Coach Blatt

### Task 5
Test the hypothesis that **number of blocks (BLK)** are from the same distribution in both the NBA playoffs and in the NBA regular season for 2015/2016 seaon for the **Toronto Raptors**. 

- We will be working with two samples with different sizes again.

```python
# create df with blocks data for the Raptors during the regular season
df_reg = df2015[df2015['TEAM_ABBREVIATION'] == 'TOR']['BLK']
df_reg
```

```python
# create df with blocks data for the Raptors during the playoffs
df_playoffs = playoffs[playoffs['TEAM_ABBREVIATION'] == 'TOR']['BLK']
df_playoffs.head()
```

```python
# Perform Welch's t-test since the sample sizes are different. Null hypothesis: number of blocks during 
# regular season and playoffs have the same mean
t_val_tor, p_val_tor = st.ttest_ind(df_reg, df_playoffs, equal_var=False)
print('The t-statistic is:', t_val_tor, '\nThe p-value is:', p_val_tor)
```

Since the p-value is below our critical value of 0.05, we reject the null hypothesis. That means the arithmetic mean of blocks during the regular season and playoffs for the Raptors is different.

### Task 6
Test the hypothesis that the number of points (PTS) scored by Cleveland Cavaliers is equally distributed for all 3 seasons. 

- In this case, we need a hypothesis test that compares more than 2 distributions at the same. (You will need to go through the scipy documentation or google to figure out what kind of test is required.)

```python
# For more than two groups, we can perform an f-test or ANOVA test. The method is st.f_oneway:
# scipy.stats.f_oneway(*samples, axis=0)
# samples must be of the same size
```

```python
# get the PTS data for each season
pts_2013 = df2013[df2013['TEAM_ABBREVIATION'] == 'CLE']['PTS']
pts_2014 = df2015[df2015['TEAM_ABBREVIATION'] == 'CLE']['PTS']
pts_2015 = df2015[df2015['TEAM_ABBREVIATION'] == 'CLE']['PTS']
pts_2015.head()
```

```python
# Perform the f-test or ANOVA test to get the p-value
f_stats, p_value = st.f_oneway(pts_2013, pts_2014, pts_2015)
print('The p-value of the ANOVA test with 3 samples is:', p_value)
```

Since the p-value is below our critical value of 0.05, we reject the null hypothesis that the PTS scored by the Cavs are equally distributed for all 3 seasons
